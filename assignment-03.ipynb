```python
#Removing Outliers
q1=data1.quantile(0.25)
q3=data1.quantile(0.75)
IQR=q3-q1
```

```python
data1=data1[~((data1<(q1-1.5*IQR)) | (data1>(q3+1.5*IQR)))]
```

```python
data1.info()
```

```python
data1['price'].fillna(data1['price'].mean(),inplace=True)
data1['area'].fillna(data1['area'].mean(),inplace=True)
data1['bedrooms'].fillna(data1['bedrooms'].mean(),inplace=True)
data1['bathrooms'].fillna(data1['bathrooms'].mean(),inplace=True)
data1['stories'].fillna(data1['stories'].mean(),inplace=True)
data1['parking'].fillna(data1['parking'].mean(),inplace=True)
```

```python
#one hot encoding
data1_main=pd.get_dummies(data1,columns=['furnishingstatus','guestroom','basement','mainroad','hotwaterheating','airconditioning'])
```

```python
data1_main.head()
```

```python
data1_main.corr()
```

```python
#split data
y=data1_main['price']
```

```python
y.head()
```

```python
X=data1_main.drop(columns=['price'],axis=1)
```

```python
X.head()
```

```python
#scale independent variables
from sklearn.preprocessing import MinMaxScaler
```

```python
name=X.columns
name
```

```python
scale=MinMaxScaler()
```

```python
X_scaled=scale.fit_transform(X)
X_scaled
```

```python
X=pd.DataFrame(X_scaled,columns=name)
X
```

```python
#split to test and train
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#test the model 
y_pred=lr.predict(X_test)
y_pred
```

```python
#measure performance
from sklearn.metrics import r2_score
acc=r2_score(y_pred,y_test)
acc
```
